```python
import boto3
import sagemaker
from sagemaker.processing import ScriptProcessor
from sagemaker import get_execution_role
from sagemaker import image_uris

role = "arn:aws:iam::711387130895:role/processor"
region = "eu-central-1"
sagemaker_session = sagemaker.Session()

# Corrected S3 paths
s3_target_model_path = "s3://YOUR_BUCKET/extracted/uniform/mnist/DP_uniform_mnist/models/"  # The root folder containing 'nodes/' and 'orchestrator/'  CHANGE ACCORDINGLY
s3_attack_model_path = "s3://YOUR_BUCKET/attack_models/"  # Bucket containing attack models # CHANGE ACCORDINGLY
s3_datasets_path = "datasets/extracted//datasets/uniform/mnist/MNIST_8_dataset/" # CHANGE ACCORDINGLY
target_models_bucket = "YOUR_BUCKET"# CHANGE ACCORDINGLY
target_models_prefix = "extracted//uniform/mnist/DP_uniform_mnist/models/" # CHANGE ACCORDINGLY
# Corrected output path
output_data_uri = f"{s3_target_model_path}/attack_results/"


# Use the pre-built Hugging Face PyTorch image
framework_version = "4.26.0"  # Choose the version of the framework you need
pytorch_image_uri = image_uris.retrieve(
    framework="huggingface",
    region=region,
    version=framework_version,
    image_scope="training",
    base_framework_version="pytorch1.13.1"  # Specify the PyTorch version you want
)
print(f"Using Hugging Face PyTorch Image: {pytorch_image_uri}")
processor = ScriptProcessor(
    image_uri=pytorch_image_uri,
    role=role,
    instance_count=1,
    instance_type="ml.m5.large",
    command=["python3"],
)

processor.run(
    code="launch_attack_mnist.py",
    arguments=[
        "--attack_bucket", "YOUR_BUCKET",# CHANGE ACCORDINGLY
        "--attack_model_prefix", "attack_models",# CHANGE ACCORDINGLY
        "--dataset_bucket",  "YOUR_BUCKET",# CHANGE ACCORDINGLY
        "--dataset_prefix", s3_datasets_path,
        "--batch_size", str(32),
        "--target_models_prefix", target_models_prefix,
        "--target_models_bucket", target_models_bucket,
        "--results_bucket", "YOUR_BUCKET",# CHANGE ACCORDINGLY
        "--results_prefix", "attack_results/uniform/with_DP/mnist/",# CHANGE ACCORDINGLY
    ],
    outputs=[sagemaker.processing.ProcessingOutput(source="/opt/ml/processing/output", destination=output_data_uri)],
    logs=True,
    wait=True,
)

print("Processing job launched successfully!")
```
